```python
def removing_extremes(df):
    res=pd.DataFrame()
    for key,subdf in df.groupby('Selling_Price'):
        m=np.mean(subdf.Selling_Price)
        sd=np.std(subdf.Selling_Price)
        reduced=subdf[(subdf.Selling_Price>=(m-sd)) & (subdf.Selling_Price<=(m+sd))]
        res=pd.concat([res,reduced],ignore_index=True)
    return res
```
